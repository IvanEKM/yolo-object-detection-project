```python
# ============================================
# Проект: распознавание объектов с помощью YOLO
# Среда: Google Colab
# Идея: быстрый прототип визуального мониторинга
# ============================================

# 1) Устанавливаем библиотеку Ultralytics (YOLO)
# Она содержит готовые предобученные модели для детекции объектов
!pip install ultralytics -q

# 2) Импортируем класс модели YOLO
from ultralytics import YOLO

# 3) Скачиваем тестовое изображение (городская сцена)
# На нём есть люди/транспорт — удобно демонстрировать распознавание объектов
!wget https://ultralytics.com/images/bus.jpg -O input.jpg

# 4) Загружаем предобученную модель YOLO
# yolov8n.pt — лёгкая и быстрая версия (подходит для демо)
model = YOLO("yolov8n.pt")

# 5) Запускаем детекцию объектов на изображении
# save=True  -> сохраняем результат с рамками
# conf=0.25  -> порог уверенности (можно увеличить до 0.4, если рамок слишком много)
results = model("input.jpg", save=True, conf=0.25)

# 6) Пишем, где лежит результат
print("Готово! файл с результатом здесь: runs/detect/predict/input.jpg")
```
